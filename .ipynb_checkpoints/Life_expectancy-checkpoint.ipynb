# MERGE WHO VS LIFE VERBOSE

```python
import pandas as pd
df_le = pd.read_csv('Resources/Life_Expectancy_Data.csv')
df_le.head()
```

```python
df_le.columns = df_le.columns.str.replace(' ', '_')
```

```python
df = pd.read_csv('Resources/lifeexpectancy-verbose.csv')
df.head()
```

```python
df.drop(columns=['index','GhoCode','PublishStateCode','YearCode','RegionCode','WorldBankIncomeGroupGroupCode','CountryCode','SexCode'], axis='columns', inplace=True)
df
```

```python
df.rename(columns={'YearDisplay':'Year','CountryDisplay':'Country'}, inplace=True)
```

```python
df_merged_wv= pd.merge(df, df_le, on=['Country','Year'], how='inner')
df_merged_wv
```

```python
df_merged_wv.columns
```

```python
df_merged_wv.shape
```

```python
df_merged_wv.isna().sum()
```

```python
df_merged_wv.dropna(inplace=True)
```

```python
df_merged_wv.shape
```

```python
for col in df_merged_wv:
    print(df_merged_wv[col].value_counts())
```

# MERGE DRINKS VS LIFE VERBOSE

```python
drinks = pd.read_csv('Resources/drinks.csv')
drinks.head()
```

```python
drinks.shape
```

```python
df = pd.read_csv('Resources/lifeexpectancy-verbose.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['index','GhoCode','PublishStateCode','YearCode','RegionCode','WorldBankIncomeGroupGroupCode','CountryCode','SexCode'], axis='columns', inplace=True)
df
```

```python
df.rename(columns={'YearDisplay':'Year','CountryDisplay':'country'}, inplace=True)
```

```python
df_merged_dv= pd.merge(df,drinks, on='country', how='inner')
df_merged_dv
```

```python
df_merged_dv.shape
```

```python
df_merged_dv.isna().sum()
```

```python
df_merged_dv.dropna(inplace=True)
```

```python
df_merged_dv
```

```python
df_merged_dv.shape
```

```python
df_merged_dv.Year.value_counts()
```

```python
df_merged_dv.to_csv('merged_DV.csv')
```
